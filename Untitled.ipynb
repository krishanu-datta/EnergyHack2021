```python
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
CALIFORNIA = pd.read_csv('CALIFORNIA.csv', encoding='utf-8')
CAROLINAS = pd.read_csv('CAROLINAS.csv', encoding='utf-8')
CENTRAL = pd.read_csv('CENTRAL.csv', encoding='utf-8')
FLORIDA = pd.read_csv('FLORIDA.csv', encoding='utf-8')
MID_ATLANTIC = pd.read_csv('MID_ATLANTIC.csv', encoding='utf-8')
MIDWEST = pd.read_csv('MIDWEST.csv', encoding='utf-8')
NEW_ENGLAND = pd.read_csv('NEW_ENGLAND.csv', encoding='utf-8')
NEW_YORK = pd.read_csv('NEW_YORK.csv', encoding='utf-8')
NORTHWEST = pd.read_csv('NORTHWEST.csv', encoding='utf-8')
ROCKY_MOUNTAINS = pd.read_csv('ROCKY_MOUNTAINS.csv', encoding='utf-8')
SOUTHEAST = pd.read_csv('SOUTHEAST.csv', encoding='utf-8')
SOUTHWEST = pd.read_csv('SOUTHWEST.csv', encoding='utf-8')
TENNESSEE = pd.read_csv('TENNESSEE.csv', encoding='utf-8')
TEXAS = pd.read_csv('TEXAS.csv', encoding='utf-8')
```

```python
regions = [CALIFORNIA, CAROLINAS, CENTRAL, FLORIDA, MID_ATLANTIC, MIDWEST, NEW_ENGLAND, NEW_YORK, NORTHWEST, ROCKY_MOUNTAINS, SOUTHEAST, SOUTHWEST, TENNESSEE, TEXAS]
USA = pd.DataFrame()
USA = USA.append(regions)

```

```python
## pd.set_option("display.max_rows", None, "display.max_columns", None)
##USA =  USA[(USA["Aggregation level"] == "County") & (USA["Unit of measure"] == "emissions (tons)") & (USA["Pollutant"] == "CO2")] 
USA =  USA[(USA["Aggregation level"] == "County") & (USA["Unit of measure"] == "percent") & (USA["Pollutant"] == "CO2")] 
USA["Annual Emissions"] = USA["January"] + USA["February"] + USA["March"] + USA["April"] + USA["May"] + USA["June"] + USA["July"] + USA["August"] + USA["September"] + USA["October"] + USA["November"] + USA["December"]
USA.sort_values(by="Annual Emissions", ascending=True)
```
